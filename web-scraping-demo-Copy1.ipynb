```python
final_df=pd.DataFrame()
for j in range(1,501):
    url='https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    web_page=requests.get(url,headers=headers).text
    soup=BeautifulSoup(web_page,'lxml')
    companies=soup.find_all('div',class_ ='companyCardWrapper')
    name=[]
    rating=[]
    review=[]
    info=[]
    for i in companies:
      name.append(i.find('h2').text.strip())
      review.append(i.find(class_ ='companyCardWrapper__ActionCount').text.strip())
      rating.append(i.find_all(class_ ='companyCardWrapper__companyRatingValue')[0].text.strip())
      info.append(i.find(class_='companyCardWrapper__interLinking').text.strip())
    d={'NAME':name,'RATING':rating,'REVIEW':review,'ABOUT':info}
    df=pd.DataFrame(d)
    final_df=pd.concat([final_df,df],ignore_index=True)
    
```
